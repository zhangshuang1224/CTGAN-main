```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('./examples/csv/Doris(2).xlsx', engine='openpyxl')
df = pd.DataFrame(df)
# 设置matplotlib支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号
```

```python
# 设置matplotlib支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 创建图形和子图
fig, ax = plt.subplots(figsize=(12, 8))

# 绘制柱状图
bars = ax.bar(df['查询条件/优化方法'], df['分散存储'], color='skyblue', align='center')

# 设置y轴的间隔，从0开始，间隔为2000
ax.set_ylim(0, max(df['分散存储']) + 2000)
ax.set_yticks(range(0, int(max(df['分散存储'])) + 2000, 2000))

# 设置x轴标题和标签（保持水平且不倾斜）
ax.set_title('分散存储')
ax.set_xlabel('查询条件/优化方法')
ax.set_ylabel('时间 (ms)')

# 设置x轴刻度位置和标签（使用换行符实现换行展示）
ax.set_xticks(df.index)
ax.set_xticklabels(df['查询条件/优化方法'], rotation=0, ha='center')  # rotation=0表示标签水平展示

# 在每个柱状条上方显示对应的y轴值
for bar, yval in zip(bars, df['分散存储']):
    ax.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval}ms', ha='center', va='bottom')

# 调整图表布局以避免标签重叠或被裁剪
plt.tight_layout()

# 显示图表
plt.show()
```
